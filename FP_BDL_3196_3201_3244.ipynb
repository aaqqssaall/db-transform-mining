##Create Transformation Class


```python
class transformation:
    def __init__(self, mongo_uri='mongodb+srv://bdlif10:bdlAMIKOMif10@bdlif10.vexcmfi.mongodb.net/?retryWrites=true&w=majority', sql_db_name='myanimelist.sqlite'):
        sqlite_conn = sqlite3.connect('myanimelist.sqlite')
        sqlite_cur = sqlite_conn.cursor()
        mongodb_client = pymongo.MongoClient(mongo_uri)
        self.sqlite_conn = sqlite_conn
        self.sqlite_cur = sqlite_cur
        self.mongodb_client = mongodb_client
        print("Objek class berhasil dibuat")
    
    def sqlite_to_df(self, tabel_name):
        query_getdata = "select * from" + " " + tabel_name
        df_from_sql = pd.read_sql_query(query_getdata, self.sqlite_conn)
        print('Transformasi '+tabel_name+' ke DataFrame berhasil')
        return df_from_sql
    
    def mongodb_insert(self, database_mongo_name, collection_name, dict_insert):
        db = self.mongodb_client[database_mongo_name]
        collection = db[collection_name]
        collection.insert_one(dict_insert)
```

##Begin Transform Database

```python
transform_to_MongoDB = transformation()
```

```python
df_info = transform_to_MongoDB.sqlite_to_df('mal_info')
df_genres = transform_to_MongoDB.sqlite_to_df('mal_genres')
df_clubs = transform_to_MongoDB.sqlite_to_df('mal_clubs')
df_date = transform_to_MongoDB.sqlite_to_df('mal_date')
df_score = transform_to_MongoDB.sqlite_to_df('mal_score')
df_watch = transform_to_MongoDB.sqlite_to_df('mal_watch')
```

```python
# initializing rem_val
rem_val = ''

def rem_vals(ele):
  global rem_val
  key, val = ele
  return val != rem_val
```

```python
for row in range(df_info.shape[0]):
  main_dict = df_info.loc[row].to_dict()
  main_dict['date'] = df_date.loc[row].to_dict()
  main_dict['score'] = df_score.loc[row].to_dict()
  main_dict['watch'] = df_watch.loc[row].to_dict()

  try:
      main_dict['clubs'] = list(map(int, df_clubs['clubs'].loc[row].split('|')))
      main_dict['genres'] = df_genres['genres'].loc[row].split('|')
      main_dict['studios'] = df_info['studios'].loc[row].split('|')
      main_dict['date']['season'] = main_dict['date']['season'].split(' ')[0]
      main_dict['date']['start_date'] = main_dict['date']['start_date'].split(' ')[0]
      main_dict['date']['end_date'] = main_dict['date']['start_date'].split(' ')[0]
  except:
      pass

  filtered_dict = {}
  for key, val in main_dict.items():
    if isinstance(val, dict):
      filtered_dict[key] = dict(filter(rem_vals, val.items()))
    else:
      filtered_dict[key] = val

  filtered_dict = {key:val for key, val in filtered_dict.items() if ((val != '')&(val != ['']))}
  transform_to_MongoDB.mongodb_insert(database_mongo_name, collection_name, filtered_dict)
```

##Preprocessing Data

###Normalize Column

```python
#mengubah bentuk query json ke menjadi selevel dalam dataframe
df = pd.json_normalize(query_list, max_level=1)
df.head(3)
```

###Check Null Values and Drop Column

```python
#menampilkan informasi data bernilai NaN (kosong)
df.isna().sum()
```

```python
#drop kolom bernilai NaN pada kolom synopsis, num_episodes, start_date
df = df.dropna(subset=['synopsis', 'num_episodes', 'date.start_date'])
df.isna().sum()
```

###Fill Column 'date.season' and Add Column 'date.start_date.year'

```python
#mengisi date.season berdasarkan bulan pada date.start_date
def fill_season(month):
  if (month in [3,4,5]):
    return 'Spring'
  elif (month in [6,7,8]):
    return 'Summer'
  elif (month in [9,10,11]):
    return 'Fall'
  elif (month in [12,1,2]):
    return 'Winter'

# for row in df.loc[df['date.season'].isna()].index:
df['date.season'] = df['date.start_date'].str[5:7:1].astype(int).apply(fill_season)

df['date.season'].isna().sum()
```

```python
def get_year(date):
  try:
    return int(date[0:4:1])
  except:
    pass

df['date.start_date.year'] = df['date.start_date'].apply(get_year)
```

###Fill NaN data with 0

```python
#mengisi data lain yang bernilai NaN dengan 0
df = df.fillna(0)
df.isna().sum()
```

##Data Mining Process

###Genre per 5 Year

```python
year_min = df['date.start_date.year'].min()
year_max = df['date.start_date.year'].max()

genres = []
year_range = []

for a in range(year_min,year_max,5):
  b = a + 5
  for row in df.loc[(df['date.start_date.year'] >= a) & (df['date.start_date.year'] < b) & (df['genres'] != 0)].index:
    for value in df['genres'][row]:
      year_range.append(str(a)+'-'+str(b-1))
      genres.append(value)

df_genres_period = pd.DataFrame(
    {'period':year_range,
     'genres':genres}).groupby(['period','genres']).size().to_frame().pivot_table(values=0, index='period', columns='genres').reset_index().rename_axis(None, axis=1)

df_genres_period = df_genres_period.fillna(0)
df_genres_period 
```

```python
min_count_row = 0.05*df_genres_period.sum(numeric_only=True).sum()
min_count_column = df_genres_period.sum(axis=0, numeric_only=True).mean()

df_genres_period = df_genres_period.loc[df_genres_period.sum(axis=1, numeric_only=True) >= min_count_row]
df_genres_period.drop([col for col, val in df_genres_period.sum(numeric_only=True).iteritems() if val <= min_count_column], axis=1, inplace=True)
df_genres_period
```
